```python
# ===============================
# 1. IMPORT LIBRARIES
# ===============================
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# 2. LOAD DATASET (IMPORTANT: use sep=';')
# ===============================
df = pd.read_csv("student-mat.csv", sep=';')

# ===============================
# 3. EXPLORE DATA
# ===============================
print("First 5 rows:")
print(df.head())

print("\nShape of dataset:")
print(df.shape)

print("\nData types:")
print(df.dtypes)

print("\nMissing values:")
print(df.isnull().sum())

# ===============================
# 4. CLEAN DATA
# ===============================
# Remove duplicates
df = df.drop_duplicates()

print("\nShape after removing duplicates:")
print(df.shape)

# ===============================
# 5. ANALYSIS
# ===============================

# Average final grade
avg_grade = df['G3'].mean()
print("\nAverage Final Grade:", avg_grade)

# Students scoring above 15
above_15 = df[df['G3'] > 15]
print("Students scoring above 15:", len(above_15))

# Correlation between study time and grade
correlation = df['studytime'].corr(df['G3'])
print("Correlation (studytime vs G3):", correlation)

# Gender performance
gender_avg = df.groupby('sex')['G3'].mean()
print("\nAverage grade by gender:")
print(gender_avg)

# ===============================
# 6. VISUALIZATIONS
# ===============================

# Histogram
plt.figure()
plt.hist(df['G3'], bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Grades (G3)")
plt.ylabel("Number of Students")
plt.show()

# Scatter Plot
plt.figure()
plt.scatter(df['studytime'], df['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()

# Bar Chart
plt.figure()
gender_avg.plot(kind='bar')
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```
